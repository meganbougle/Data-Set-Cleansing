```python
import pandas as pd

df= pd.read_csv('Airbnb_Open_Data.csv')
```

```python
print(df.head())
```

```python
print(df.describe)
```

```python
df.info()
print(df)
```

```python
print(df.isnull().sum())
```

```python
#dropeamos la columna "license" ya que tiene muchos valores nulos y solo dos valores no nulos (no aporta informacion)
df.drop('license', axis=1, inplace=True)
print(df.head())
```

```python
df.drop('country code', axis=1, inplace=True)
print(df.head())
```

```python
df["service fee"] = df["service fee"].str.replace("$", "").str.replace(",", "").astype(float)
df["price"] = df["price"].str.replace("$", "").str.replace(",", "").astype(float)
```

```python
#LLenando valores nulos
df['house_rules'].fillna('None', inplace=True)
df['country'].fillna('United States', inplace=True)
df['lat'].fillna(0, inplace=True)
df['long'].fillna(0, inplace=True)
df['host_identity_verified'].fillna('unconfirmed', inplace=True)
df['instant_bookable'].fillna('FALSE', inplace=True)
df['number of reviews'].fillna(0, inplace=True)
df['minimum nights'].fillna(1, inplace=True)
df['availability 365'].fillna(0, inplace=True)
df['price'].fillna(0, inplace=True)
df['service fee'].fillna(0, inplace=True)
df['reviews per month'].fillna(0, inplace=True)
df['calculated host listings count'].fillna(0, inplace=True)
df['review rate number'].fillna(0, inplace=True)
df.fillna('None', inplace=True)

print(df.isnull().sum())
```

```python
#separando las fechas en año y mes
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')

df['year_review'] = df['last review'].dt.year
df['month_review'] = df['last review'].dt.month

df['year_review'] = df['year_review'].fillna(0).astype(int)
df['month_review'] = df['month_review'].fillna(0).astype(int)

print(df[['last review', 'year_review', 'month_review']].head())

```

```python
df.drop('last review', axis=1, inplace=True)
```

```python
for column in df.columns:
    print(f"Unique values in '{column}':")
    print(df[column].unique())
    print()
```

```python
df['neighbourhood group'] = df['neighbourhood group'].replace({'manhatan': 'Manhattan', 'brookln': 'Brooklyn'})
print(df['neighbourhood group'].unique())
```

```python
df['minimum nights'] = pd.to_numeric(df['minimum nights'], errors='coerce')
df['minimum nights'] = df['minimum nights'].clip(lower=1, upper=365)
print(df['minimum nights'])
```

```python
df['availability 365'] = pd.to_numeric(df['availability 365'], errors='coerce')
df['availability 365']=df['availability 365'].clip(lower=0, upper=365)
print(df['availability 365'])
```

```python
manhattan_neighborhoods = ['East Harlem', 'Upper East Side', 'Upper West Side', 'Chelsea', 'Greenwich Village', 'Chinatown','East Village', 'Washington Heights', 'Harlem']
brooklyn_neighborhoods = ['Williamsburg', 'Bedford-Stuyvesant', 'Bushwick', 'Bedford', 'Prospect Heights', 'Clinton Hill']
bronx_neighborhoods = ['Eastchester']
queens_neighborhoods = ['Woodside', 'Queens Village']


def neighborhoodgrp(row):
    if row['neighbourhood'] in manhattan_neighborhoods and row['neighbourhood group'] == 'None':
        return 'Manhattan'
    elif row['neighbourhood'] in brooklyn_neighborhoods and row['neighbourhood group'] == 'None':
        return 'Brooklyn'
    elif row['neighbourhood'] in bronx_neighborhoods and row['neighbourhood group'] == 'None':
        return 'Bronx'
    elif row['neighbourhood'] in queens_neighborhoods and row['neighbourhood group'] == 'None':
        return 'Queens'
    return row['neighbourhood group']


df['neighbourhood group'] = df.apply(neighborhoodgrp, axis=1)
```

```python
df.columns = df.columns.str.strip().str.upper().str.replace(' ', '_')
columns_to_upper = [
    'NAME', 
    'HOST_IDENTITY_VERIFIED', 
    'HOST_NAME', 
    'NEIGHBOURHOOD_GROUP', 
    'NEIGHBOURHOOD', 
    'COUNTRY', 
    'INSTANT_BOOKABLE', 
    'CANCELLATION_POLICY', 
    'ROOM_TYPE', 
    'HOUSE_RULES'
]

for col in columns_to_upper:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.upper()
```

```python
print(df)
```

```python
df.to_csv('Airbnb_Cleaned.csv', index=False)
```
