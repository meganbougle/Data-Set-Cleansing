```python
columnExcluida=df['duration']
df.fillna('None', inplace=True)
df['duration']=columnExcluida
print(df)
```

```python
df.info()
print(df)
```

```python
#Limpiando espacios en blanco
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
print(df)
```

```python
#separando las fechas en año y mes
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

df['year_added'] = df['year_added'].fillna(0).astype(int)
df['month_added'] = df['month_added'].fillna(0).astype(int)

print(df[['date_added', 'year_added', 'month_added']].head())
```

```python
df = df.drop(columns=['date_added'])
```

```python
for column in df.columns:
    print(f"Unique values in '{column}':")
    print(df[column].unique())
    print()
```

```python
wrong_ratings = df[df['rating'].str.contains(r'^\d+ min$', na=False)]
print(wrong_ratings)
```

```python
df.loc[wrong_ratings.index, 'duration'] = df.loc[wrong_ratings.index, 'rating']
df.loc[wrong_ratings.index, 'rating'] = None  # Set 'rating' to None for these rows

# Step 4: Standardize the 'rating' column by replacing 'UR' with 'NR'
df['rating'] = df['rating'].replace({'UR': 'NR'})
print(df['rating'].unique())
```

```python
unique_countries = set()
unique_genres = set()
df['country'].dropna().apply(lambda x: unique_countries.update(x.split(',')))
df['listed_in'].dropna().apply(lambda x: unique_genres.update(x.split(',')))
print(sorted(unique_countries))
print(sorted(unique_genres))
```

```python
def replace_countries(entry):
    if isinstance(entry, str):
        countries = [country.strip() for country in entry.split(',')]
        updated_countries = []
        for country in countries:
            if country == 'Soviet Union':
                updated_countries.append('Russia')
            elif country in ['West Germany', 'East Germany']:
                updated_countries.append('Germany')
            else:
                updated_countries.append(country)
        return ', '.join(updated_countries)
    return entry


df['country'] = df['country'].apply(replace_countries)
print(df['country'].unique())
```

```python
def extract_duration(duration):
    if isinstance(duration, str):
        
        parts = duration.split()
        if parts and parts[0].isdigit():
            return int(parts[0])
    return None

```

```python
df['duration_minutes'] = df.apply(lambda x: extract_duration(x['duration']) if x['type'] == 'Movie' else None, axis=1)
df['duration_seasons'] = df.apply(lambda x: extract_duration(x['duration']) if x['type'] == 'TV Show' else None, axis=1)
```

```python
print(df.head())
```

```python
df = df.drop(columns=['duration'])
df['duration_minutes'].fillna('NA', inplace=True)
df['duration_seasons'].fillna('NA', inplace=True)
print(df.head())
```

```python
df.to_csv('netflix_titles_cleaned.csv', index=False)
```
